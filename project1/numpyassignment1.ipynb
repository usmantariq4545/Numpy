```python
import numpy as np
import matplotlib.pylab as plt
import os
from PIL import Image
import glob
images = []
resized = []
resized_array = []
for file in glob.glob (r'\Users\STAR\OneDrive\Desktop\images\*.jfif'):

    img = Image.open(file)
    images.append(img)
for image in images:
    res = image.resize([200,200])
    resized.append(res)
for i in resized:
    numpy = np.array(i,dtype="uint8")
    resized_array.append(numpy)
    print(resized_array, '\n',np.shape(resized_array))
```
